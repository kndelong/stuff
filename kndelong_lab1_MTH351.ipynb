```r
boxplot(weight.m$Weight, weight.f$Weight,
  main = "Weights of Males & Females",
  ylab = "Weight in lbs.",
  names = c("Males", "Females")
  )
```

In the boxplot comparison above we see that the average weight for males is ~10lbs heavier than females in the class. We can also see that the range for both sets of data is almost equivalent.
